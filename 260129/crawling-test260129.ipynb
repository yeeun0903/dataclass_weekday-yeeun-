```python
# 크롤링 사이트에서 텍스트 타이틀만 찾아오기

import requests
from bs4 import BeautifulSoup

url = "https://davelee-fun.github.io/trial/board/news.html"

res = requests.get(url)
soup = BeautifulSoup(res.text, "html.parser")

items = soup.select("div.list_item.symph_row")
for item in items : 
    title = item.select_one("span.subject_fixed")
    if title != None :
        print(title.get_text(" ", strip=True))
```

```python
# 찾아온 텍스트 타일에 순번 매기기 - 1

import requests
from bs4 import BeautifulSoup

url = "https://davelee-fun.github.io/trial/board/news.html"

res = requests.get(url)
soup = BeautifulSoup(res.text, "html.parser")

items = soup.select("div.list_item.symph_row")
for index, item in enumerate(items) : 
    title = item.select_one("span.subject_fixed")
    if title != None :
        print(index + 1, title.get_text(" ", strip=True))
    
```
