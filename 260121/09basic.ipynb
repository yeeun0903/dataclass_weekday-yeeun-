# 파이썬 Class
- 클래스 : 자주 반복적으로 사용하고자 하는 기능을 메뉴얼처럼 정의해놓고, 필용한 상황에 따라서 해당 값을 복제해서 가져가서 사용하기 위한 일종의 문법 체계

# 프로그래밍 언어의 역사
- C, 파스카, 코볼, 포트란 : 1990년 이전에 나온 언어 -> 클래스 없음
  > 절차지향 언어 : PP(Procedural Programming)
  → 냉장고에 코끼리를 넣는 방법
  → 냉장고 문을 연다
  → 코끼리를 냉장고 안에 넣는다
  → 냉장고 문을 닫는다
  → 직관, 명확 => 절차

- JS, TS, Javam C++, C#, Python : 모두 클래스 있음
  > 객체지향 언어 : OOP(Object Oriented Programming)
  → 코끼리라는 대상의 정보를 메뉴얼(클래스)로 만든다
  → 해당 메뉴얼을 기반으로 코끼리1(객체)를 만든다
  → 냉장고라는 대상의 정보를 메뉴얼(클래스)로 만든다
  → 해당 메뉴얼을 기반으로 냉장고1(객체)를 만든다
  → 코끼리1 객체가 가지고 있는 기능을 사용해서 냉장고1 객체의 문을 열게한다
  → 코끼리1 객체가 가지고 있는 기능을 사용해서 냉장고1 객체안으로 들어가게 한다
  → 코끼리1 객체가 가지고 있는 기능을 사용해서 냉장고1 객체의 문을 닫게한다

```python
class Student :
    def __init__(self, name, grade) : 
        self.name = name
        self.grade = grade

    def study(self) : 
        print(self.name, "공부합니다.")

    def take_exam(self, subject):
        print(self.name, subject, "공부합니다.")
```

```python
student1 = Student("David", 3)
student2 = Student("Jane", 2)

print(student1.name)
print(student1.grade)
print(student2.name, student2.grade)
```

```python
student1.study()
```

```python
student1.take_exam("수학")
```
