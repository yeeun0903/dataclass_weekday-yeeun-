```python
# urllib + bs4 => 비교적 옛날 방식

from urllib.request import urlopen
from bs4 import BeautifulSoup

url = "https://davelee-fun.github.io"
response = urlopen(url)
html = response.read().decode("utf-8")

soup = BeautifulSoup(html, "html.parser")

featured = soup.select_one(".featured-posts")
items = featured.select(".card-text")

for item in items :
    print(item.get_text(" ", strip=True))
```

```python
# HTTP (Hyper Text Tranfer Protocol) 상태코드
# 2xx
# 3xx
# 4xx
# 5xx

# 1xx (정보): 요청을 받았으며 프로세스를 계속한다
# 2xx (성공): 요청을 성공적으로 받았으며 인식했고 수용하였다
# 3xx (리다이렉션): 요청 완료를 위해 추가 작업 조치가 필요하다
# 4xx (클라이언트 오류): 요청의 문법이 잘못되었거나 요청을 처리할 수 없다
# 5xx (서버 오류): 서버가 명백히 유효한 요청에 대해 충족을 실패했다

# 서버 측에서 응답메세지를 의도적으로 차단, 오류발생
# -> 예외 조항처리

# 크롤링 > 못함 > Why? > 오타? or 서버?

import requests
from bs4 import BeautifulSoup

url = "https://davelee-fun.github.io"
headers={
    "User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)
if response.status_code != 200 :
    print("페이지 없음")
else :
    soup = BeautifulSoup(response.content, "html.parser") 
    data = soup.select(".card-text")
    for item in data:
        print(item.get_text(" ", strip=True))
```

```python
# 복수의 다중페이지 크롤링 하는 방법
# 1) 각 페이지별 주소창에서 주소값을 확인
# 2) 페이별 주소값의 차이점 = 패턴 확인
# 3) 확인한 패턴에 부합하는 파이썬 코드를 작성해 페이지별 크롤링을 진행한다
```

```python
# https://davelee-fun.github.io
# https://davelee-fun.github.io/page2/
# https://davelee-fun.github.io/page3/
# https : 프로토콜(통신규약)
# davelee-fun.github.io : 도메인 주소
# /page3/ : 파라미터 => 경로를 정의
# /index.html : 기본페이지 => 없어도 됨

import requests
from bs4 import BeautifulSoup

url = "https://davelee-fun.github.io"
headers={
    "User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)

if response.status_code != 200 :
    print("페이지 없음")
else :
    # 8회의 반복!! -> 0부터 7까지 반복 : 8회
    for page_num in range(8) :
        if page_num == 0 :
            response = requests.get("https://davelee-fun.github.io", headers=headers)
        else :
                response = requests.get(f"https://davelee-fun.github.io/page{str(page_num + 1)}", headers=headers)
        soup = BeautifulSoup(response.content, "html.parser") 
        data = soup.select(".card-text")
        for item in data:
            print(item.get_text(" ", strip=True))
```

```python
import openpyxl

# 데이터 수집 : 50행
# 데이터 저장준비 : 40행
# 데이터 저장 : 30행
# 120행 -> ctrl + enter -> 에러
# 1개의 코드에서 모두 진행 X
# 리팩토링 (refactoring)

def write_excel_template(filename, sheetname, listdata) :
    excel_file = openpyxl.Workbook()
    excel_sheet = excel_file.active
    excel_sheet.column_dimensions["A"].width = 100
    
    if sheetname != "" :
        excel_sheet.title = sheetname
    for item in listdata :
        excel_sheet.append(item)
        
    excel_file.save(filename)
    excel_file.close()
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://davelee-fun.github.io"
headers={
    "User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)
product_lists = list()

if response.status_code != 200 :
    print("페이지 없음")
else :

    # 8회의 반복!! -> 0부터 7까지 반복 : 8회
    for page_num in range(8) :
        if page_num == 0 :
            response = requests.get("https://davelee-fun.github.io", headers=headers)
        else :
                response = requests.get(f"https://davelee-fun.github.io/page{str(page_num + 1)}", headers=headers)
        soup = BeautifulSoup(response.content, "html.parser") 
        data = soup.select(".card")
        for item in data : 
            product_name = item.select_one(".card-text")
            product_date = item.select_one(".post-date")
            product_info = [product_name.get_text(" ", strip=True), product_date.get_text(" ", strip=True)]
            product_lists.append(product_info)
```

```python
write_excel_template("tmp.xlsx", "상품정보", product_lists)
```
