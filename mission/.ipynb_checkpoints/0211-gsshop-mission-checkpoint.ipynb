```python
# https://www.gsshop.com/ 크롤링  > 티비편성표 > 오늘 > 방송시간대 + 방송상품
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
import time

URL = "https://www.gsshop.com/" 
service = Service(ChromeDriverManager().install())
options = Options()

# options.add_argument("--headless")
# options.add_argument("--disable-gpu")
# options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=1920,1080")
options.add_argument("--start-maximized")
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36")
options.add_argument("--lang=ko_KR")
options.add_argument("--no-sandbox")

# GS SHOP 사이트 접속하기
driver = webdriver.Chrome(service=service, options=options)                     
driver.get(URL) 
time.sleep(2)

# TV편성표 버튼 클릭
tv_btn = driver.find_element(By.ID, "tvScheduleMain").click()
time.sleep(2)

# 오늘 버튼 클릭
today_btn = driver.find_element(By.ID, "day_20260211").click()
time.sleep(2)

# 스크롤 내리기?
driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
time.sleep(2)

# 방송시간대 + 방송상품 찾아오기 
items = driver.find_elements(By.CSS_SELECTOR, "article.items")

for idx, item in enumerate (items) :
    time = item.find_element(By.CSS_SELECTOR, "span.times").text.strip()
    product = item.find_element(By.CSS_SELECTOR, "dt.prd-name").text.strip()
    print(f"{idx+1} {time} : {product}")

driver.quit()
```
