```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
import time

URL = "https://www.ssg.com/" 
service = Service(ChromeDriverManager().install())
options = Options()

# options.add_argument("--headless")
# options.add_argument("--disable-gpu")
# options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=1920,1080")
options.add_argument("--start-maximized")
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36")
options.add_argument("--lang=ko_KR")
options.add_argument("--no-sandbox")

# SSG 사이트 접속하기
driver = webdriver.Chrome(service=service, options=options)                     
driver.get(URL) 
time.sleep(2)

# 신세계몰 버튼 클릭 
new_btn = driver.find_element(By.XPATH, "//*[@class='gnb_mall_list']/li[2]/a").click()
time.sleep(2)

# 쓱특가 버튼 클릭 
ssg_btn = driver.find_element(By.XPATH, "//*[@class='smgnb_navbar']/div[2]/ul/li[1]/a").click()
time.sleep(2)

# 리빙 버튼 클릭
living_btn = driver.find_element(By.ID, "tabs-:Rcledaal6nj6:--tab-7").click()
time.sleep(2)

# 스크롤 내리기
driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
time.sleep(2)

# 50개 상품 : 상품명 + 판매금액 + 할인률 + URL
items = driver.find_elements(By.CSS_SELECTOR, "div.css-1kmkxxv")
print(items)
for idx, item in enumerate (items) :
    product = item.find_element(By.CSS_SELECTOR, "p.chakra-text.css-19bfb2ae").text.strip()
    price = item.find_element(By.CSS_SELECTOR, "em.css-aywnvu").text.strip()
    promotion = item.find_element(By.CSS_SELECTOR, "span.css-idkz9h").text.strip()
    URL = item.find_element(By.CSS_SELECTOR, "a.chakra-link.css-1j5cjm").get_attribute("href")
    
    print(f"{idx+1} {product}/{price}/{promotion}/{URL}")

driver.quit()
```
