### 문자열 관련 파이썬 문법

1. 문자열에 있는 특정 문자 갯수 세기 : count()
2. 문자열에 있는 특정 문자의 위치 파악 : index() -> 반드시 있어야 하는 값을 찾아와라!
3. 문자열에 있는 특정 문자의 위치를 파악 (에러를 발생 x) : find() -> -1
4. 서로 다른 문자열을 하나로 합치고자 할 때 : join()
5. 문자열에서 앞,뒤 공백을 없앨 때 : strip() // rstrip() : 오른쪽 공백 제거 // lstrip() : 왼쪽 공백 제거
                                  -> 인자값이 없으면 기본적으로 공백을 없앰
7. 대,소문자 변경 : upper() : 대문자화 , lower(): 소문자화
8. 문자열을 쪼개고자 할 때 : split() -> 리스트를 변환 -> 인자값이 없으면 기본적으로 공백을 기준으로 나눔
9. 문자열을 찾아서 원하는 문자열을 교체하고자 할 때 : replace(인자값 => 대상, 무엇으로 교체할 지)

```python
data = "Dave David"
data.count("D")
```

```python
string = "Dave ID is Dave"
string.index("D")
```

```python
string = "dave ID is Dave"
string.index("D")
```

```python
string = "dave ID is Dave"
string.find("f")
```

```python
string = "12345"
comma = ","

comma.join(string) #",".join(string)
"|".join(string)
```

```python
data = "   Dave "
data.strip()
```

```python
data = "   Dave   "
# data.lstrip()
data.rstrip()
```

```python
string= "         9999999999999(DAVE)8888888888  "
string.strip("98() ") # 여백을 없애고 싶으면 띄어쓰기 하면 됨!
```

```python
string = "Dave"
# string.upper()
string.lower()
```

```python
string = "Dave goes to Korea"
string.split()[3] # 구분자(구분하는 기준)를 아무것도 넣지 않는다면, 공백을 구분자로 선택함!
```

```python
string = "Dave/goes to/Korea"
string.split("/")
```

```python
string2 = "(David)"
string2.replace("(", "").replace(")", "") # method chaining 기법
```
