```python
import requests # 데이터를 수집해오는 크롤러
import bs4 # 찾아온 데이터를 보다 쉽게 이해하고, 해석할 수 있도록 아름답게 바꿔주는 역할

url = "https://news.daum.net/tech"
headers = {
    "User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36"
}

# 함수와 메서드함수 구분해서 지칭!!
response = requests.get(url, headers=headers)
response.encoding = "utf-8" # 한국어 -> 컴퓨터가 이해할 수 있는 언어로 변환
soup = bs4.BeautifulSoup(response.text, "html.parser")

articles = soup.select("ul.list_newsheadline2 li a.item_newsheadline2")

for article in articles :
    title = article.text.strip()
    link = article["href"]
    print(f"{title} - {link}")
```

```python
# 기상청 홈페이지 - 내 버전

import requests 
import bs4 

url = "https://www.weather.go.kr/w/forecast/overall/short-term.do"
headers = {
    "User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36"
}


response = requests.get(url, headers=headers)
response.encoding = "utf-8" 
soup = bs4.BeautifulSoup(response.text, "html.parser")

paragraph = soup.select("div.cmp-view-content > p.summary")

for p in paragraph :
    print(p.get_text(" ", strip=True))
```
