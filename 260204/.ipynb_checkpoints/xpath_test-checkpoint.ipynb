```python
"""
/ : 절대경로 -> /html/body/div/div
// : 특정 문서상에서 해당되는 내용을 모두 다 선택 -> //h1 : 문서상 h1태그로 시작되는 값은 다 찾아와라
//*[@href] : 현재 문서에서 href 속성을 가지고 있는 모든 태그를 찾아와라
//a[@href="https://google.com"] : 현재 문서 내 https://google.com 해당 주소값을 가지고 있는 href 속성을 가지고 있는 모든 a를 찾아와라
(//a)[3] : 현재 문서 내 모든 a 태그를 찾아와라, 하지만 3번째 등장하는 a만 찾아와라 (단, 0부터 x, 1부터 시작!)
(//table)[last()] : 현재 문서 내 모든 table 태그를 찾아와라, 하지만 마지막번째만 찾아와라
(//a)[position() < 3] : 현재 문서 내 모든 a 태그를 찾아와라, 하지만 3보다 미만인 a만 찾아와라 (1, 2번째까지만)
(//div[@*]) : 문서상에서 div 태그를 모두 찾아라, 하지만 속성이 하나라도 존재하는 div 태그를 찾아와라 (속성이 무엇이든 상관없음)
contains(@class, "속성") and contains(@class, "속성") -> 복수의 속성을 가지고 있는 경우에 한함
"""
# / 직계자손선택자
# // 후손선택자 -> 자손선택자 = all
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from webdriver_manager.chrome import ChromeDriverManager
from urllib.request import urlretrieve
import time

# 상수값으로서의 변수 목록
URL = "https://moolix.netlify.app/" # 특이성이 있는, 고정하는, 변하지 않는 변수는 대문자로 표시하기도 함

service = Service(ChromeDriverManager().install())
options = Options()

# options.add_argument("--headless")
# options.add_argument("--disable-gpu")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=1920,1080")
options.add_argument("--start-maximized")
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36")
options.add_argument("--lang=ko_KR") 
options.add_argument("--no-sandbox") 

driver = webdriver.Chrome(service=service, options=options)

driver.get(URL)
time.sleep(3)

# 1. xpath -> 검색바안에 placeholder -> 영화검색제목이라는 텍스트 크롤링
# search_input = driver.find_element(By.XPATH, "//input[@placeholder='영화 검색 (제목)']") 
# print("검색창 placeholder", search_input.get_attribute("placeholder"))

# 2. 각 영화의 제목만 수집해서 출력
# titles = driver.find_elements(By.XPATH, "//div[@class='title']") 
# titles = driver.find_elements(By.XPATH, "//section/article[@class='card']//div[@class='title']")
# titles = driver.find_elements(By.XPATH, "//*[@id='results']/article//div[@class='title']")
# titles = driver.find_elements(By.XPATH, "//section/article[@class='card']//div[contains(@class, 'title')]")



# 3. 각 영화의 이미지 주소 (속성)를 수집하고 출력
# links = driver.find_elements(By.XPATH, "//section/article[@class='card']//img") 
# for link in links :
#     print(link.get_attribute('src'))

# 4. 사이트 내 첫번재 영화(하우스 메이드) 클릭 후 상세페이지 접속 > 제목, 개봉연도, 줄거리
# driver.find_element(By.XPATH, "//section/article[@class='card']//img[1]").click() 
driver.find_element(By.XPATH, "//article[contains(@class, 'card')]").click() 
time.sleep(2)

title = driver.find_element(By.XPATH, "//h2").text  
date_and_rating = driver.find_element(By.XPATH, "//div[@class='muted']").text 
date_text = date_and_rating.split("·")[0].strip()
plot = driver.find_element(By.XPATH, "//div[@class='stack']/div[1]").text
# plot = driver.find_element(By.XPATH, "//div[@class='stack']/div/following-sibling::div[1]").text
print("영화제목 :", title)
print("개봉연도 :", date_text)
print("줄거리 :", plot)



driver.quit()
```

```python
# 0204 미션
# XPATH 방식으로 인기언어 1위~10위까지 크롤링 후 출력하기

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from webdriver_manager.chrome import ChromeDriverManager
from urllib.request import urlretrieve
import time

URL = "https://www.tiobe.com/tiobe-index/" 

service = Service(ChromeDriverManager().install())
options = Options()

# options.add_argument("--headless")
# options.add_argument("--disable-gpu")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=1920,1080")
options.add_argument("--start-maximized")
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36")
options.add_argument("--lang=ko_KR") 
options.add_argument("--no-sandbox") 

driver = webdriver.Chrome(service=service, options=options)

driver.get(URL)
time.sleep(2)
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(2)
 
names = driver.find_elements(By.XPATH, "//tbody//tr//td[5]")
for name in names[:10] :
    print(name.text)
time.sleep(2)    


driver.quit()
```
