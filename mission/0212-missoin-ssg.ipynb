```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
import time

URL = "https://www.ssg.com/" 
service = Service(ChromeDriverManager().install())
options = Options()

# options.add_argument("--headless")
# options.add_argument("--disable-gpu")
# options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=1920,1080")
options.add_argument("--start-maximized")
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36")
options.add_argument("--lang=ko_KR")
options.add_argument("--no-sandbox")

# SSG 사이트 접속하기
driver = webdriver.Chrome(service=service, options=options)                     
driver.get(URL) 
time.sleep(2)

# 신세계몰 버튼 클릭 
new_btn = driver.find_element(By.XPATH, "//*[@class='gnb_mall_list']/li[2]/a").click()
time.sleep(2)

# 쓱특가 버튼 클릭 
ssg_btn = driver.find_element(By.XPATH, "//*[@class='smgnb_navbar']/div[2]/ul/li[1]/a").click()
time.sleep(2)

# 식품 버튼 클릭
food_btn = driver.find_element(By.ID, "tabs-:Rcledaal6nj6:--tab-8").click()
time.sleep(2)

# 스크롤 내리기
driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
time.sleep(2)

# 상품명，할인，가격，주소 -> 다른 건 다 있는데!!! 할인률이 없는 상품들이 섞여있음 -> try 구문 써야징
items = driver.find_elements(By.CSS_SELECTOR, "div.template-grid-item.css-8kawlz")[:50]
for idx, item in enumerate(items):
    # 상품명
    try:
        name = item.find_element(By.XPATH, ".//p").text.replace("\n", " ").strip()
    except:
        name = ""
    # 할인률
    try:
        sale = item.find_element(By.XPATH, ".//em[1]").text.replace("\n", " ").strip()
    except:
        sale = ""
    #가격
    try:
        price = item.find_element(By.XPATH, ".//em[2]").text.replace("\n", " ").strip()
    except:
        price = ""
    # 주소
    try:
        url = item.find_element(By.XPATH, ".//a[@href]").get_attribute("href")
    except:
        url = ""
    print(f"{idx+1} {name}\n {sale}, {price} {url}")
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
import time

URL = "https://www.ssg.com/" 
service = Service(ChromeDriverManager().install())
options = Options()

# options.add_argument("--headless")
# options.add_argument("--disable-gpu")
# options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=1920,1080")
options.add_argument("--start-maximized")
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36")
options.add_argument("--lang=ko_KR")
options.add_argument("--no-sandbox")

# 1. SSG 사이트 접속하기
driver = webdriver.Chrome(service=service, options=options)                     
driver.get(URL) 
time.sleep(2)

# 2. shinsegaemall_tab
shinsegaemall_tab = driver.find_element(By.XPATH, "//a[contains(@class, 'gnb_mall_link') and contains(text(), '신세계몰')]")
shinsegaemall_tab.click()

# 3. ssg_tab
ssg_tab = driver.find_element(By.XPATH, "//a[contains(@class, 'menu_lnk clickable') and contains(text(), '쓱-특가')]")
ssg_tab.click()

# 4. food_btn
# food_btn = driver.find_element(By.XPATH, "//button[@data-index='8']")
# food_btn.click()
food_btn = driver.find_element(By.ID, "tabs-:Rcledaal6nj6:--tab-8").click()
time.sleep(2)
# 5. 50개 아이템 
# 5-1 상품들을 감싸안고 있는 요소 확인 
# 5-2 해당 요소 안에 실제 상품별 요소 확인 
# 5-3 각 상품별 요소 안, 상품명/할인률/판매금액/url 링크 요소 확인
# 이미 5-2 에서 진행하는 단계 : 리스트 + for 반복문 쓰자

# 개별상품 크롤링 코드
results = list() # results= []

while len(results) < 50 : 
    cards = driver.find_elements(By.CSS_SELECTOR, "div.template-grid-item.css-8kawlz")
    
    for card in cards :
        try :
            a = card.find_element(By.CSS_SELECTOR, "a[href]")
            url = a.get_attribute("href")
        except :
            continue
        try :
            name = card.find_element(By.CSS_SELECTOR, "p.chakra-text.css-19bfb2a > span.chakra-text.css-5oz1h7").text.strip()
        except :
            name = ""
        results.append({"상품명": name, "URL": url})

driver.quit()

results
```
