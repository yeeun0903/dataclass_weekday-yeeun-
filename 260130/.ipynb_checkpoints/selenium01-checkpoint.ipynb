```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time 

# 로컬 컴퓨터의 브라우저 드라이버와  동일한 버전으로 설치 & 객체생성
service = Service(ChromeDriverManager().install())

# 가상 브라우저가 실제처럼 보일 수 있도록 설정하는 옵션
options = Options()
options.add_argument("--headless") # -> 가상 브라우저가 띄어졌음에도 우리 눈에는 안 나타나도록
options.add_argument("--disable-gpu") # 굳이 가상 브라우저를 우리 눈에 보이지 않게했다면, gpu 작동 불필요, headless와 세트 둘다 있거나 없거나
options.add_argument("--disable-dev-shm-usage") # 가상 메모리 사용하지 말라
options.add_argument("--window-size=1920,1080") # 가상 브라우저의 사이즈 정보 설정
options.add_argument("--start-maximized") # 특정 사이트 반응형 -> 데스크탑에 맞춰라!
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36")
options.add_argument("--lang=ko_KR") # 브라우저를 사용하는 해당 국가의 모국어
options.add_argument("--no-sandbox") # sandbox : 규제 -> 크롤링 자체를 정책적으로 차단하는 가이드,메뉴얼을 피해 우회 

# 브라우저 정보값을 가지고 있는 객체를 매개변수로 받아서 크롤링 준비완료
driver = webdriver.Chrome(service=service)  # 크롬 브라우저를 통해서
driver.get("https://davelee-fun.github.io/")

elements = driver.find_elements(By.CSS_SELECTOR, "div.wrapthumbnail img")

for element in elements :
    print(element.get_attribute("src"))

driver.quit()

# requests.get() -> 특정 url 주소에 방문해서 값을 찾아와라
```

### selenium 문법
0. find(단일 태그) / findall(복수 태그) // select(복수의 선택자: . #) / select_one(단일 선택자)
1. find_elements : 복수의 값을 찾아온다 -> 리스트 (동일한 요소를 전체 다 찾아서 가져온다)
2. find_element : 단수의 값을 찾아온다 -> 위에서 문서를 읽다가, 최초로 발견한 값을 찾아오고 종료
3. By : .TAG_NAME (태그를 중심), .ID (아이디중심) -> 문서 1번밖에 사용불가, .CSS_SELECTOR (선택자중심 -> 클래스, 태그, 아이디), .NAME
4. get_attribute() : 특정 속성의 속성값을 찾아오고자 할 때
5. .clear() : 특정 영역 내 값을 삭제 (Form 요소 기준) > imput[type=radio, check, tect], textarea 외
6. .send_keys() : 어떤 값 혹은 키보드의 키값을 전달
7. Keys : 키보드의 키를 입력하고자 할 때
8. sleep(N) : N초의 시간동안을 지연시키고자 할 때
9. screenshot(filename) : filename으로 스크린 캡처 가능
10. set_window_size(가로, 세로) : 윈도의 창을 가로 세로 사이즈로 조정

requests : html, css, js -> 사이트 문서를 다 받아놓은 후 해당 문서를 읽고서 그 안에서 원하는 요소를 찾아오는 방식
selenium : 현재 브라우저에 출력한 데이터를
