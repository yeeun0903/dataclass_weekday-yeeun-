https://books.toscrape.com/ -> 메인 페이지에서 “책 제목+가격” 20개 가져오기
-> 가산점 : 재고여부 
-> 책을 클릭했을 때, 이동할 수 있는 url 
-> url 이동 후, product description

```python
import requests
from bs4 import BeautifulSoup

url = "https://books.toscrape.com/"

res = requests.get(url)
soup = BeautifulSoup(res.text, "html.parser")

base_url = "https://books.toscrape.com/"

items = soup.select("article.product_pod")

for item in items : 
    title = item.select_one("h3")
    price = item.select_one("p.price_color")
    stock = item.select_one("p.instock.availability")
    link = item.select_one("a") 
    total_url = base_url + link["href"]
    
    print(title.get_text(" ", strip=True), 
          price.get_text(" ", strip=True), 
          stock.get_text(" ", strip=True),
          total_url
         )
    

```
