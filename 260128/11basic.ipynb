### 정규표현식
1. 단순 개발 업무 -> 정규표현식 인지 // 많이 사용하는 일 x
2. 데이터 다루는 직무 -> 정규표현식
- 어떤 문자열 속에서 내가 원하는 문자를 추출해오는데 있어서 탁월
- GA4, BigQuery, MySQL > SQL
- 데이터 및 정보의 시각화 : Tableau, PowerBI, LookerStidio -> 정규표현식

- 패턴중심 : 일관성을 가지고 있는 형태 패턴

1) [0-9] : 숫자를 찾아오겠다 -> \d = decimal = 10진수
2) ^[0-9] : ^ = 캐럿 = 무조건 ~으로 시작한다는 의미
3) [^0-9] : 부정 = not -> 숫자가 아닌 것을 찾아라 = \D
4) [\t\n\r\f\v] : t : tab => 긴여백 / n : new line => 개행 : 행바꿈 / r : carriage return / f : form feed / v : vertical tab
5) [^\t\n\r\f\v] : white space (페이지, 탭여백, 행간, 개행) 문자가 아닌 것들만 찾아와라
6) [A-Z] : 영문대문자
7) [a-z] : 영문소문자
8) [A-Za-z] : 영문대소문자
9) [A-Za-z0-9] : 영문대소문자 혹은 숫자
10) [가-힣] : 한글
11) [가-힣A-Za-z0-9]
12)  . : 문자 1개를 의미함 -> D.A -> D + / | Z | E | A ... + A
13) re.sub(패턴, 바꿀데이터(어떻게 바꿀건지), 원본데이터)
14)  ? : 물음표를 중심으로 앞에 있는 문자가 0번 또는 1번 표시되는 패턴 (gold14k) (gold) -> 없을 수도 있고 1번 있거나
15)  * : * 를 중심으로 앞에 있는 문자가 0번 또는 그 이상 (1번의 제약 x) [A-Z]*
16)  + : + 를 중심으로 앞에 있는 문자가 1번 또는 그 이상 (0번을 허용하지 않음)
17) {n} : 앞에 있는 문자가 n번 반복되는 패턴을 찾아라 "AD{2}A" -> ADDA
18) {m, n} : 앞에 있는 문자가 m번 반복되는 패턴부터 n번 반복되는 패턴까지 (범위)를 찾아라 - 범위를 지정할 때, 띄어쓰기 불가!!
19) [] : 대괄호 안에 포함된 문자는 허용하겠다는 패턴
20) - : 범위를 지정함 [a-d] [d-t]
21) search() : 왼쪽부터 패턴에 매칭되는 문자를 최초로 찾는 순간 정지 / findall() : 문자열 모두를 끝까지 찾음 -> 리스트 형식으로 값을 반환

```python
import requests
from bs4 import BeautifulSoup
import re

url = "https://crawlingtest.netlify.app/"
res = requests.get(url)
res.raise_for_status

soup = BeautifulSoup(res.text, "html.parser")
datas = soup.select("#dev_course_list .course")

for data in datas :
    text = data.get_text(" ", strip=True)
    clean_text = re.sub(r"\[[0-9]+\]", "", text) # escape sequence = \n, \r, \t
    print(clean_text)
```

```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://davelee-fun.github.io/blog/crawling_stock_example.html")
soup = BeautifulSoup(res.text, "html.parser")

data = soup.select("ul.table-stock.sty1 > li.row_sty")

# 전前처리 -> 데이터수집 -> 분석을 효율적으로 하기위한 데이터 정리 (값이 비어있음 x, 특수문자, 숫자, 문자, 화이트스페이스) -> 분석 목적
for item in data :
    company = item.select_one("div.st_name > span.name").get_text(" ", strip=True).replace(" ", "").replace("\n","")
    price = item.select_one("div.st_price > span.price").get_text(" ", strip=True).replace(" ", "").replace("\n","")
    rate = item.select_one("div.st_rate > span.icon").get_text(" ", strip=True).replace(" ", "").replace("\n","")
    print(company, price, rate)
```
