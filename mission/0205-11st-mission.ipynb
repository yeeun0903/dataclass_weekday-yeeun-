```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
import time

URL = "https://www.11st.co.kr/?gad_source=1&gad_campaignid=20978125333&gclid=Cj0KCQiAnJHMBhDAARIsABr7b86ImtZHE-QYvXXedOj52wK9RhHiDL06nL-9D6wgBj0AM_wTXuP2kzUaAnjzEALw_wcB&gbraid=0AAAAADQ_R8Id2G6g1iLrgK_DHAtYxg5U3&utm_term=11+%B9%F8%B0%A1&utm_campaign=%B1%B8%B1%DBPC_11%B9%F8%B0%A1&utm_source=%B1%B8%B1%DB_PC_S&utm_medium=%B0%CB%BB%F6" 
service = Service(ChromeDriverManager().install())
options = Options()

# options.add_argument("--headless")
# options.add_argument("--disable-gpu")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=1920,1080")
options.add_argument("--start-maximized")
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36")
options.add_argument("--lang=ko_KR") 
options.add_argument("--no-sandbox") 

driver = webdriver.Chrome(service=service, options=options)
driver.get(URL)
time.sleep(3)

element = driver.find_element(By.CLASS_NAME, "search_text")
element.click()
element.send_keys("캠핑용품")
element.send_keys(Keys.RETURN)
time.sleep(3)

products = driver.find_elements(By.CSS_SELECTOR, "div.c-search-list li.c-search-list__item a > span.sr-only")

for index, product in enumerate(products[:100]):
    print(index + 1, product.text)
    
driver.quit()
```
