```python
import requests

url = "https://apis.data.go.kr/B551011/GoCamping/basedList"
request_params = {
    "numOfRows" : "10",
    "pageNo" : "1",
    "MobileOS" : "ETC",
    "MobileApp" : "camping",
    "serviceKey" : "bb54fc1f25d5b537cfbc5d1430aa31b14cc8f65a0710615096749046f79033e7",
    "_type" : "xml"
}

#params= 이것도 headers= 처럼 고정! 문법!
response = requests.get(url, params=request_params)
print(response)
```

```python
# 만약 위의 방식이 안될 때!

import subprocess
import xml.etree.ElementTree as ET

url = "https://apis.data.go.kr/B551011/GoCamping/basedList?numOfRows=50&pageNo=1&MobileOS=ETC&MobileApp=camping&serviceKey=bb54fc1f25d5b537cfbc5d1430aa31b14cc8f65a0710615096749046f79033e7&_type=xml"

result = subprocess.run(["curl", "-k", url], capture_output=True, text=True, encoding="utf-8")
root = ET.fromstring(result.stdout)

print(root.findall(".//item"))
```
