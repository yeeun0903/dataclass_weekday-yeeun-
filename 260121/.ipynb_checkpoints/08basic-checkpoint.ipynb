```python
price = "100 달러"
exchange_rate = 1480

# price라는 변수에 있는 100 달러의 값을 환율을 적용해서 한화로 변환해서 값을 출력하시오.

# 문자열 : 인덱스 -> 슬라이싱

K_won = int(price[0:3]) * exchange_rate
print(K_won)
print(f"{K_won:,}원") # 1000의 단위마다 ','을 찍어라
```

```python
# 반복문을 활용해서 구구단 2단 ~ 9단까지 아래에 출력하시오.

# 2 ~ 9단 반복

for index_1 in range(2, 10) : 
    print(f"==== {index_1}단 ====")
    for index_2 in range(1, 10) :
        print(f"{index_1} x {index_2} = {index_1 * index_2}")
    print("\n") # escape sequence => 문자처럼 생겼지만, 문자이상의 고유한 문법을 가지고 있는 문자
```
